```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as img
img_name=r'F:\screenshots\Screenshot_20190113-120028_Chrome.jpg'
#img_name= np.random.randn(10, 10)
im_data=img.imread(img_name)
#print(im_data)
#print(np.shape(im_data))
red_data=im_data[:,:,0]
blue_data=im_data[:,:,1]
green_data=im_data[:,:,2]
cmap='hot'
plt.figure(figsize=(15,10))
plt.imshow(red_data,cmap='nipy_spectral', interpolation="sinc")
#plt.set_cmap('nipy_spectral')
plt.title('image')
plt.colorbar(orientation='horizontal')
plt.axis('off')
#print(im_data)
#print(np.shape(im_data))
new_img=np.array([blue_data,green_data,red_data])
new_img=np.swapaxes(new_img,0,1)
new_img=np.swapaxes(new_img,1,2)
#print(np.shape(new_img))
#new_img.thumbnail((64, 64), Image.ANTIALIAS)
cmap='gray'
plt.figure(figsize=(15,10))
plt.imshow(new_img,cmap=cmap,interpolation='nearest',alpha=0.8)
plt.title('new_image')
plt.axis('off')
plt.colorbar(ticks=[1,10,20,30,190],orientation='horizontal')
plt.show()


image2=np.array([1.0*green_data,0.5*blue_data,0.5*red_data])
image2=np.swapaxes(image2,0,1)
image2=np.swapaxes(image2,1,2)
cmap='jet'
plt.figure(figsize=(5,10))
plt.imshow(red_data,cmap=cmap,interpolation='nearest')
plt.title('red new_image')
plt.axis('off')
plt.show()
cmap='jet'
plt.figure(figsize=(5,10))
plt.imshow(green_data,cmap=cmap,interpolation='nearest')
plt.title('green new_image')
plt.axis('off')
plt.show()
cmap='jet'
plt.figure(figsize=(5,10))
plt.imshow(blue_data,cmap=cmap,interpolation='nearest')
plt.title('blue new_image')
plt.axis('off')
plt.show()
plt.figure(figsize=(5,10))
plt.hist(red_data.ravel(),bins=100,fc='red',ec='k')
plt.show()
plt.figure(figsize=(5,10))
plt.hist(blue_data.ravel(),bins=100,fc='blue',ec='k',range=(0,10.0))
plt.show()
```
