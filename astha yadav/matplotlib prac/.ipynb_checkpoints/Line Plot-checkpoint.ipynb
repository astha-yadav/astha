```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,5)
y=x**3
plt.plot(x, y,'ko-',label='numbers',linewidth=5,marker='^',markeredgecolor='red',alpha=0.9,markerfacecolor='yellow',markeredgewidth=0.9,
        markerfacecoloralt='black',animated=True,linestyle='-')
plt.title('line plot')
plt.grid(True)
plt.xlabel('----numbers---')
plt.ylabel('----exponential----')
plt.show()
```
