```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 5])
plt.ylabel('some numbers')
plt.xlabel('some values')
plt.title('yess')
plt.show()
```

```python
plt.plot([1, 2, 3, 5],[4,7,8,9],'ro')
plt.ylabel('some numbers')
plt.xlabel('some values')
plt.title('yess')
plt.axis([0,8,0,15])
plt.show()
```

```python
import numpy as  np
t=np.arange(0,5,0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
plt.scatter('a', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
