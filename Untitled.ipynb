```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
## Read
img = cv2.imread("20190530_110717.jpg")
final_image=img
## convert to hsv
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
## mask of green (36,25,25) ~ (86, 255,255)
# mask = cv2.inRange(hsv, (36, 25, 25), (86, 255,255))
mask = cv2.inRange(hsv, (36, 25, 25), (90, 255,255))
## slice the green
imask = mask>0
green = np.zeros_like(img, np.uint8)
green[imask] = img[imask]
## save
cv2.imwrite("green3.png", green)
plt.imshow(green,interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()


im_gray = cv2.imread('green3.png', cv2.IMREAD_GRAYSCALE)
(thresh, im_bw) = cv2.threshold(im_gray, 128, 255,cv2.THRESH_OTSU)
plt.imshow(im_bw,interpolation = 'bicubic',shape = (10,10))
#plt.grid(True)
plt.show()




for i in range(1):
    im_floodfill = im_bw.copy()
 
    # Mask used to flood filling.
    # Notice the size needs to be 2 pixels than the image.
    h, w = im_bw.shape[:2]
    mask = np.zeros((h+2, w+2), np.uint8)
 
    # Floodfill from point (0, 0)
    cv2.floodFill(im_floodfill, mask, (0,0), 255);
 
    # Invert floodfilled image
    im_floodfill_inv = cv2.bitwise_not(im_floodfill)
 
    # Combine the two images to get the foreground.
    im_out = im_bw | im_floodfill_inv
 
    # Display images.
    plt.imshow(im_bw,shape = (10,10))
    final_image=im_bw
    #plt.imshow(im_floodfill,interpolation = 'bicubic',shape = (10,10))
    #plt.imshow(im_floodfill_inv,interpolation = 'bicubic',shape = (10,10))
    plt.imshow(im_out,interpolation = 'bicubic',shape = (10,10))
    plt.show()
   
    ret,thresh1 = cv2.threshold(im_out,15,255,cv2.THRESH_BINARY) #the value of 15 is chosen by trial-and-error to produce the best outline of the skull
    kernel = np.ones((5,5),np.uint8) #square image kernel used for erosion
    erosion = cv2.erode(thresh1, kernel,iterations = 1) #refines all edges in the binary image
    opening = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel) #this is for further removing small noises and holes in the image
    plt.imshow(closing, 'gray',shape = (10,10)) #Figure 2
    plt.xticks([]), plt.yticks([])
    plt.show()




for i in range(10):
    im_floodfill =closing.copy()
 
    # Mask used to flood filling.
    # Notice the size needs to be 2 pixels than the image.
    h, w = closing.shape[:2]
    mask = np.zeros((h+2, w+2), np.uint8)
 
    # Floodfill from point (0, 0)
    cv2.floodFill(im_floodfill, mask, (0,0), 255);
 
    # Invert floodfilled image
    im_floodfill_inv = cv2.bitwise_not(im_floodfill)
 
    # Combine the two images to get the foreground.
    im_out = closing | im_floodfill_inv
 
    # Display images.
    plt.imshow(closing,shape = (10,10))
    final_image=closing
    #plt.imshow(im_floodfill,interpolation = 'bicubic',shape = (10,10))
    #plt.imshow(im_floodfill_inv,interpolation = 'bicubic',shape = (10,10))
    plt.imshow(im_out,interpolation = 'bicubic',shape = (10,10))
    plt.show()
   
    ret,thresh1 = cv2.threshold(im_out,15,255,cv2.THRESH_BINARY) #the value of 15 is chosen by trial-and-error to produce the best outline of the skull
    kernel = np.ones((5,5),np.uint8) #square image kernel used for erosion
    erosion = cv2.erode(thresh1, kernel,iterations = 1) #refines all edges in the binary image
    opening = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel) #this is for further removing small noises and holes in the image
    plt.imshow(closing, 'gray',shape = (10,10)) #Figure 2
    plt.xticks([]), plt.yticks([])
    plt.show()



contours, hierarchy = cv2.findContours(closing,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cv2.imshow('cleaner', closing)
cv2.drawContours(closing, contours, -1, (255, 255, 255), 4)
#ret,thresh = cv2.threshold(im_gray,127,255,0)
#contours,hierarchy = cv2.findContours(thresh, 1, 2)
cnt = contours[0]
M = cv2.moments(cnt)
print(M)
areas = []
for contour in contours:
  ar = cv2.contourArea(contour)
  areas.append(ar)
max_area = max(areas)
max_area_index = areas.index(max_area)
print(max_area_index)
cnt = contours[max_area_index]
print(cnt)




cv2.drawContours(closing, [cnt], 0, (255, 255, 255), 3, maxLevel = 0)
plt.imshow(closing, 'gray',shape = (10,10))
plt.xticks([]), plt.yticks([])
plt.show()
area = cv2.contourArea(cnt)
print(area)
cv2.imwrite("green2.png", closing)





area_in_cm2=area/20647.55
area_in_cm2
```
