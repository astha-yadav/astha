```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
fig.suptitle('yeah title')
fig,ax_lst=plt.subplots(2,2)
```

```python
x = np.linspace(0, 2, 100)
plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quad")
plt.plot(x,x**3,label="cubic")
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
def my_plotter(ax, d1, d2,param_dict):
    out=ax.plot(d1, d2,**param_dict )
    return out

d1, d2, d3, d4 = np.random.randn(4,100)
fig,ax = plt.subplots(1, 1)
my_plotter(ax, d1, d2, {'marker' : 'x'})
```

```python
def my_plotter(ax, d1, d2,param_dict):
    out=ax.plot(d1, d2,**param_dict )
    return out

d1, d2, d3, d4 = np.random.randn(4,100)
fig,(ax1,ax2) = plt.subplots(1, 2)
my_plotter(ax1, d1, d2, {'marker' : 'x'})
my_plotter(ax2, d1, d2, {'marker' : 'o'})
```
